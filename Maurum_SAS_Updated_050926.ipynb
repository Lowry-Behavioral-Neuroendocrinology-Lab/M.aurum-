QIIME2's Moving Pictures tutorial gives concise explainations of what each metric is:

**Alpha diversity**

- Shannon’s diversity index (a quantitative measure of community richness)

- Observed Features (a qualitative measure of community richness)

- Faith’s Phylogenetic Diversity (a qualitiative measure of community richness that incorporates phylogenetic relationships between the features)

- Evenness (or Pielou’s Evenness; a measure of community evenness)

**Beta diversity**

- Jaccard distance (a qualitative measure of community dissimilarity)

- Bray-Curtis distance (a quantitative measure of community dissimilarity)

- Unweighted UniFrac distance (a qualitative measure of community dissimilarity that incorporates phylogenetic relationships between the features)

- Weighted UniFrac distance (a quantitative measure of community dissimilarity that incorporates phylogenetic relationships between the features)

# Rarefaction

 ## **Generate alpha and beta diversity rarefaction plots for reference**

Vector at the end of the file name = one column of all the alpha diversity datapoints for each sample
Beta diversity: Distance matrix: beta diversity. The closer the distance to 1 the more dissimilar it is, if it is identical it will be closer to 0 (Unifrac). Bray Curtis/Jaccard is the old style and doesn't take into consideration the phylogenetic tree. Unweighted Unifrac and Weighted Unifrac take into consideration the phylogenetic tree and "unifrac" is the portion of the phylogenetic tree that's shared. So this measures the distance (how similar/different) each strain of bacteria is/sample is in relation to a phylogenetic tree. Kathy Louzapone created the weighted/unweighted unifrac. Jacard and unweighted unifrac don't take species abundances into account. Weighted Unifrac is the most informative about what's there and how much fo it is there. But something else ot consider is that unweighted unifrac and jacard give more value to rare species. Unweighted unifrac doesn't take species abundance into consideration (for rare bacteria to get equal representation among the ones that) it will pick up anything rare/not as much of something present. Weighted unifrac takes abundance into account. We look at both weighted and unweighted. 

Alpha Diversity: Evenness, Faith, Observed, Shannon. 
Observed OTUs: how many species do we see in the sample. Example: 3 bananas 3 apples 2 oranges so Observed OTU: 3
Shannon OTUs is similar to observed but it takes into account the quantities. Example: Calculates/taking into account the quantity as well as species. Spits out a number/calculated via equations. (Richness/Diversity)
Faith PD: Based on how many species we see and how dissimilar they are in their phylogeny. PD: Phylogenic Diversity. (RICHNESS/Diversity)
Evenness: How evenly distributed the species are. Example: 3 apples, 3 bananas, 3 oranges (these are even); 3 apples 2 bananas 1 oranges (these are not even). Dominant species. (Based on abundances)

From Moving Pictures tutorial: The visualization will have two plots. The top plot is an alpha rarefaction plot, and is primarily used to determine if the richness of the samples has been fully observed or sequenced. If the lines in the plot appear to “level out” (i.e., approach a slope of zero) at some sampling depth along the x-axis, that suggests that collecting additional sequences beyond that sampling depth would not be likely to result in the observation of additional features. If the lines in a plot don’t level out, this may be because the richness of the samples hasn’t been fully observed yet (because too few sequences were collected), or it could be an indicator that a lot of sequencing error remains in the data (which is being mistaken for novel diversity).

# ANCOM-BC- Preliminary analysis

Using the code and instructions from the microbiome workshop.

Before doing ANCOM, we need to filter out samples with fewer features than we rarefied at (<1500, as seen in core metrics) using the filter-samples method.

Additionally, we’re going to filter out low abundance ASVs (those that are observed less than 50 times total and/or in less than 4 samples). 

We do this because we need features to be present in multiple samples, otherwise we are really only testing presence/absence. Filtering also helps limit the FDR (false discovery rate), which helps prevent us from counting something as significant when it really is just noise

qiime feature-table filter-samples \
  --i-table noMito_noChloro-filtered-table.qza \
  --p-min-frequency 1500 \
  --o-filtered-table noMito_noChloro-filtered-1500-table.qza
  
qiime feature-table filter-features \
  --i-table noMito_noChloro-filtered-1500-table.qza \
  --p-min-frequency 50 \
  --p-min-samples 4 \
  --o-filtered-table noMito_noChloro-1500-table_abund.qza
  
qiime taxa collapse \
  --i-table noMito_noChloro-1500-table_abund.qza \
  --i-taxonomy silva-classified/silva-taxonomy.qza \
  --p-level 7 \
  --o-collapsed-table noMito_noChloro-1500-table_abund_L7.qza
  
Now one last thing before running ANCOM. One thing that ANCOM can’t handle is when things like extraction controls, that have no metadata, are included in your metadata file. So if you hav this in your original metadata file, create another version of the metadata that is exactly the same, it just doesn’t include the extraction control rows. 

Now we can finally run the ANCOM test. A couple of things here:
Note that, like when we ran beta diversity stats, we have to decide which variable we want to test before running the command and tell QIIME2 in the –p-formula parameter.
I mentioned before that a pseudocount (addition of 1) is required for all features. Note that although it isn’t explicit, this is happening internally 

#Run ANCOM on Condition- test between condidtion 

qiime composition ancombc \
  --i-table noMito_noChloro-1500-table_abund_L7.qza \
  --m-metadata-file Maurum_metadata_all.tsv \
  --p-formula 'condition' \
  --o-differentials ancombc_condition.qza
  
qiime composition tabulate \
  --i-data ancombc_condition.qza \
  --o-visualization ancombc_condition.qzv
  
#Run ANCOM on Treatment- test between treatment

qiime composition ancombc \
  --i-table noMito_noChloro-1500-table_abund_L7.qza \
  --m-metadata-file Maurum_metadata_all.tsv \
  --p-formula 'treatment' \
  --o-differentials ancombc_treatment.qza
  
qiime composition tabulate \
  --i-data ancombc_treatment.qza \
  --o-visualization ancombc_treatment.qzv
